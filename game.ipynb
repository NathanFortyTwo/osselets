```python
import random as rd
from numpy import transpose,array
from IPython.display import clear_output

# game of osselets
class Player():
    def __init__(self):
        self.score= 0
        self.dice = 0
        self.board = [[0,0,0],[0,0,0],[0,0,0]]

    @property
    def place_restante(self):
        res = sum([row.count(0) for row in self.board])
        return res


def place_dice(player,row_index):
    selected_row = player.board[row_index]
    try :

        place_index = selected_row.index(0) # find the first 0 in the column
    except IndexError:
        raise ValueError("Row is full")
    
    player.board[row_index][place_index] = player.dice

    
def get_score_col(row):
    score = 0
    for index,value in enumerate(row):
        count = row.count(value)
        score+= value*count # if we have 5,5,3, the score is 5*2 + 5*2 +3 = 23
    return score

class Osselets():
    def __init__(self):
        
        self.players = [Player(),Player()]
        self.turn = 0
        self.done = False
    
    def roll_dice(self):
        self.players[self.turn].dice = rd.randint(1,6)
        value = self.players[self.turn].dice
        print(f"Player {self.turn} rolled a {value}")
        return value

    def action(self,row_index):
        print("C'est au tour de : ",self.turn)
        place_dice(self.players[self.turn],row_index) # place dice
        value = self.players[self.turn].dice
        self.turn = (self.turn+1)%2 # change turn

        # if value is in the same column of the other player, replace it with 0
        row = self.players[self.turn].board[row_index]
        if value in row:
            index = row.index(value)
            self.players[self.turn].board[row_index][index] = 0
            print(f"Player {self.turn} lost a dice")


   
    def get_score(self):
        player = self.players[self.turn]
        rows = player.board 
        score = 0
        for row in rows:
            score+=get_score_col(row)
        print(f"Player {self.turn} score has {score} points")

    def display_boards(self):
        clear_output()
        for player in self.players:
            #transpose the board to display it correctly
            board = transpose(array(player.board))
            print(f"Player {self.players.index(player)} board")
            for row in board:
                print(row)
            print("")

    def is_done(self):
        return self.players[self.turn].place_restante == 0 or self.players[(self.turn+1)%2].place_restante == 0
    
    def start(self):
        while not self.done:
            value = self.roll_dice()

            placed = False
            while not placed:
                action = int(input(f"Choose a column to place your dice {value}: "))
                try:
                    self.action(action) # increments turn
                    placed = True
                except ValueError:
                    print("Column is full, place somewhere else")
                    
            self.display_boards()
            
            if self.players[self.turn].place_restante == 0:
                self.done = True
                score1 = self.get_score()
                self.turn = (self.turn+1)%2
                score2 = self.get_score()
                if score1 > score2:
                    print(f"Player {self.turn} won: {score1} > {score2}")
                elif score1 < score2:
                    print(f"Player {(self.turn+1)%2} won : {score2} > {score1}")


```

```python
if __name__ == "__main__":
    oss = Osselets()
    oss.start()


#TODO 
# sanitize user input
# add a bot
# add a gui
```
